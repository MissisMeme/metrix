## Описание датасета

* Campaign_id - ID кампании;
* Geo - геопозиция, на которую настроена рекламная кампания;
* Buyer - байер, выполнивший запуск;
* Impressions - количество показов рекламных объявлений;
* Clicks - количество кликов; 
* Installs - количество установок;
* Registrations - количество регистраций;
* Spend - Затраченная сумма;
* Revenue_ad (1d/7d/30d) - Выручка с рекламы от пользователей, зарегистрировавшихся в приложении, на 1/7/30 день их жизни
* Revenue_in_app (1d/7d/30d) - Выручка с покупок от пользователей, зарегистрировавшихся в приложении, на 1/7/30 день их жизни
* Payers (1d/7d/30d) - Количество покупателей в приложении на 1/7/30 день их жизни

```python
df = pd.read_csv('ad_camp.csv', sep=',')
df
```

```python
df.info()
```

```python
df = df.replace(',', '.', regex=True)
df.head()
```

```python
df = df.astype({'Spend' : 'float64', 'Revenue_ad1d' : 'float64', 'Revenue_in_app1d': 'float64', 'Revenue_ad7d': 'float64', 'Revenue_in_app7d': 'float64', 'Revenue_ad30d' : 'float64', 'Revenue_in_app30d' : 'float64'})
```

```python
df.info()
```

# Кампании в каких гео показали себя хуже всего?

```python
df['Geo'].value_counts()
```

```python
geo_grouped = df.groupby('Geo').agg({'Spend' : 'sum', 'Revenue_ad30d' : 'sum'})
geo_grouped['ROI'] = (geo_grouped['Revenue_ad30d'] - geo_grouped['Spend']) / geo_grouped['Spend'] * 100

geo_grouped.sort_values('ROI')
```

```python
plt.figure(figsize=(10, 7))

plt.bar(geo_grouped.index, geo_grouped['ROI'])

plt.xlabel('Geo')
plt.ylabel('ROI')
plt.title('ROI by Geo')
plt.xticks(rotation=30)

plt.show()
```

Вывод: рекламная кампания в **australia oceania** с коэффициентом ROI **-94.37** показала себя хуже всего

# Кто из медиабайеров лучше всего окупается?

```python
df['Buyer'].value_counts()
```

```python
buyer_grouped = df.groupby('Buyer').agg({'Spend' : 'sum', 'Revenue_ad30d' : 'sum'})
buyer_grouped['ROI'] = (buyer_grouped['Revenue_ad30d'] - buyer_grouped['Spend']) / buyer_grouped['Spend'] * 100

buyer_grouped.sort_values('ROI', ascending=False)
```

```python
plt.figure(figsize=(10, 7))

plt.bar(buyer_grouped.index, buyer_grouped['ROI'])

plt.ylabel('ROI')
plt.title('ROI by Buyer')

plt.show()
```

Вывод: **buyer_3** с коэффициентом ROI **4.76** окупается лучше всего

# Динамика ROI 

## Geo

```python
geo_ROI = df.groupby('Geo').agg({'Spend' : 'sum', 'Revenue_ad1d' : 'sum', 'Revenue_ad7d' : 'sum', 'Revenue_ad30d' : 'sum'})
geo_ROI['ROI_1d'] = (geo_ROI['Revenue_ad1d'] - geo_ROI['Spend']) / geo_ROI['Spend'] * 100
geo_ROI['ROI_7d'] = (geo_ROI['Revenue_ad7d'] - geo_ROI['Spend']) / geo_ROI['Spend'] * 100
geo_ROI['ROI_30d'] = (geo_ROI['Revenue_ad30d'] - geo_ROI['Spend']) / geo_ROI['Spend'] * 100

geo_ROI
```

```python
plt.figure(figsize=(10, 7))

plt.plot([1, 7, 30], [geo_ROI['ROI_1d'], geo_ROI['ROI_7d'], geo_ROI['ROI_30d']])

plt.xlabel('Days')
plt.ylabel('ROI')
plt.title('Динамика ROI by Geo')
plt.legend(geo_ROI.index)

plt.show()
```

## Buyer

```python
buyer_ROI = df.groupby('Buyer').agg({'Spend' : 'sum', 'Revenue_ad1d' : 'sum', 'Revenue_ad7d' : 'sum', 'Revenue_ad30d' : 'sum'})
buyer_ROI['ROI_1d'] = (buyer_ROI['Revenue_ad1d'] - buyer_ROI['Spend']) / buyer_ROI['Spend'] * 100
buyer_ROI['ROI_7d'] = (buyer_ROI['Revenue_ad7d'] - buyer_ROI['Spend']) / buyer_ROI['Spend'] * 100
buyer_ROI['ROI_30d'] = (buyer_ROI['Revenue_ad30d'] - buyer_ROI['Spend']) / buyer_ROI['Spend'] * 100

buyer_ROI
```

```python
plt.figure(figsize=(10, 7))

plt.plot([1, 7, 30], [buyer_ROI['ROI_1d'], buyer_ROI['ROI_7d'], buyer_ROI['ROI_30d']])

plt.xlabel('Days')
plt.ylabel('ROI')
plt.title('Динамика ROI by Buyer')
plt.legend(buyer_ROI.index)

plt.show()
```

# CAC (Цена привлечения одного клиента)



## Geo

```python
geo_CAC = df.groupby('Geo').agg({'Spend' : 'sum', 'Payers1d' : 'sum', 'Payers7d' : 'sum', 'Payers30d' : 'sum'})
geo_CAC['CAC_1d'] = geo_CAC['Spend'] / geo_CAC['Payers1d']
geo_CAC['CAC_7d'] = geo_CAC['Spend'] / geo_CAC['Payers7d']
geo_CAC['CAC_30d'] = geo_CAC['Spend'] / geo_CAC['Payers30d']

geo_CAC
```

```python
plt.figure(figsize=(10, 7))

plt.bar(geo_CAC.index, geo_CAC['CAC_30d'])

plt.xlabel('Geo')
plt.ylabel('CAC')
plt.title('CAC by Geo')

plt.show()
```

Самое дешевое привлечение за 30 дней одного клиента в Латинской Америке (44.56), а самое дорогое в Западной Европе (161.94).

```python
plt.figure(figsize=(10, 7))

plt.plot([1, 7, 30], [geo_CAC['CAC_1d'], geo_CAC['CAC_7d'], geo_CAC['CAC_30d']])

plt.xlabel('Days')
plt.ylabel('CAC')
plt.title('Динамика CAC by Geo')
plt.legend(geo_CAC.index)

plt.show()
```

В динамике CAC меняется аналогично ROI. К 7-му дню CAC уменьшается, а ROI повышался, что говорит о том, что кампании старались корректировать, но изначально, возможно, была выбрана неверная тактика. После 7-го дня резких изменений в показателях не наблюдается.

# CPC (Стоимость клика)

## Geo

```python
geo_CPC = df.groupby('Geo').agg({'Spend' : 'sum', 'Clicks' : 'sum'})
geo_CPC['CPC'] = geo_CPC['Spend'] / geo_CPC['Clicks']

geo_CPC.sort_values('CPC')
```

```python
plt.figure(figsize=(10, 7))

plt.bar(geo_CPC.index, geo_CPC['CPC'])

plt.xlabel('Geo')
plt.ylabel('CPC')
plt.title('CPC by Geo')
plt.xticks(rotation=30)

plt.show()
```

Самый дорогой клик в Австралии Океании стоимостью 0.4, а самый дешевый в Северной Африке стоимостью 0.007.

При этом дешевизна клика не говорит о высоких доходах в Северной Африке, так как ROI был -89.07 - это второй результат в худшем ROI по geo после Автралии Океании.

## Buyer


```python
buyer_CPC = df.groupby('Buyer').agg({'Spend' : 'sum', 'Clicks' : 'sum'})
buyer_CPC['CPC'] = buyer_CPC['Spend'] / buyer_CPC['Clicks']

buyer_CPC.sort_values('CPC')
```

```python
plt.figure(figsize=(10, 7))

plt.bar(buyer_CPC.index, buyer_CPC['CPC'])

plt.ylabel('CPC')
plt.title('CPC by Byuer')

plt.show()
```

Самый дорогой клик в 0.17 у самого окупаемого buyer_3. Самый дешевый клик в 0.02 у менее окупаемого buyer_5.

# RPC (Доход с каждого клика)

## Geo

```python
geo_RPC = df.groupby('Geo').agg({'Revenue_ad30d' : 'sum', 'Clicks' : 'sum'})
geo_RPC['RPC'] = geo_RPC['Revenue_ad30d'] / geo_RPC['Clicks']

geo_RPC.sort_values('RPC')
```

```python
plt.figure(figsize=(10, 7))

w = 0.4
offset = 0.2
n_ticks = np.arange(7)

plt.bar(n_ticks - offset, geo_RPC['RPC'], label='RPC', width=w)
plt.bar(n_ticks + offset, geo_CPC['CPC'], label='CPC', width=w)

plt.xlabel('Geo')
plt.ylabel('Индексы')
plt.title('Сравнение CPC и RCP')
plt.xticks(n_ticks, geo_RPC.index, rotation=30)
plt.legend(['RPC', 'CPC'])

plt.show()
```

По всем геопозициям доход с одного клика меньше, чем цена клика.


## Buyer

```python
buyer_RPC = df.groupby('Buyer').agg({'Revenue_ad30d' : 'sum', 'Clicks' : 'sum'})
buyer_RPC['RPC'] = buyer_RPC['Revenue_ad30d'] / buyer_RPC['Clicks']

buyer_RPC.sort_values('RPC')
```

```python
plt.figure(figsize=(10, 7))

w = 0.4
offset = 0.2
n_ticks = np.arange(5)

plt.bar(n_ticks - offset, buyer_RPC['RPC'], label='RPC', width=w)
plt.bar(n_ticks + offset, buyer_CPC['CPC'], label='CPC', width=w)

plt.ylabel('Индексы')
plt.title('Сравнение CPC и RCP')
plt.xticks(n_ticks, buyer_RPC.index)
plt.legend(['RPC', 'CPC'])

plt.show()
```

Только buyer_3 получил доход с одного клика больше, чем расход на него.
